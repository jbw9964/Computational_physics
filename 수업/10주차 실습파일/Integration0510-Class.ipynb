# 1.Newton-Cotes
we approximate $f(x)$ by a polymonial of degree $n$ that intersects all the nodes <br>
Lagrange's form of this polynomial:
$$ P_n(x) =   \sum_{i=0}^n f(x_i) \ell_i(x) dx $$ 

Integral:

$$I = \int_{a}^{b} P_n(x) dx =   \sum_{i=0}^n f(x_i) \int_{a}^{b}  \ell_i(x) dx = \sum_{i=0}^{n}  A_i f(x_i)$$

$$A_i=\int_{a}^{b}  \ell_i(x) dx $$

## 1-1. Trapezoidal Rule(n=1)
one panel , two points

$$I=\sum_{i=0}^{1}  A_i f(x_i)$$

# 3.Gaussian Integration

Weighting function $(\omega)$ is used to increase the accuracy of the integral result by assigning weights based on the variation of function values over the integration interval.

$$\int_a^b \omega(x)P_m(x)dx = \sum_{i=0}^n A_iP_m(x_i)$$

If we use a special set of orthogonal polynomials based on the weighting function, we can calculate the integral more easily. <br>
These orthogonal polynomials have the property of being orthogonal to the weighting function, so when we calculate the integral, we can obtain an approximate result by calculating the coefficients of each polynomial and summing them up.
