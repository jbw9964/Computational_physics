# Initial value

$$y''=-y , ~~ y(0)=1 ,~~y'(0)=0$$
동일한 x에 대한 보조조건과 미분방정식을 푸는 문제

차수 n의 상미분방정식 : $y^{(n)}=f(x,y,y' ,...,y^{(n-1)})$ 으로 표현된다.

$$y_0 = y , ~~ y_1 = y' ,~~ y_2 = y'' ,~~\dots~~ y_{n-1}=y^{(n-1)}$$

$$y'_0 = y_1 ,~~ y'_1=y_2 ,~~ y'_2=y_3 , ~~\dots~~ y'_{n-1}=y_n=f(x,y_0,y_1,...,y_n)$$

$$y'=F(x,y)= \begin{bmatrix} 
y_1 \\ 
y_2 \\ 
\vdots \\
f(x,y)
\end{bmatrix}$$ 

## Euler

초기값과 $i$번째 1차 도함수의 기울기를 이용해 i+1번째 함수값 예측


$$tan\theta = \frac{y_{i+1}-y_{i}}{x_{i+1}-x_{i}}=\frac{dy}{dx} = f(x,y)$$

$$y_{i+1}=y_i+f(x_i,y_i)(x_{i+1}-x_i)$$

$$y_{i+1}=y_i+f(x_i,y_i)h$$

## 2차 Runge-Kutta

$$y(x+h)=y(x)+c_0F(x,y)h + c_1F\Big[x+ph,y+qhF(x,y)\Big]h ~~~~~~~~~~~~~~~~~~~~(a)$$


를 다음의 Taylor 급수와 일치시킨다.

$$y(x+h)=y(x)+y'(x)h + \frac{1}{2!}y''(x)h^2$$

$$~~~~~~~~~~~~~~~~~~~~=y(x)+F(x,y)h +\frac{1}{2}F'(x,y)h^2~~~~~~~~~~~~~~~~~~~(b)$$ 

$F\big[x+ph,qhF(x,y)\big]=F(x,y)+\frac{\delta F}{\delta x}ph+qh\sum_{i=1}^{n-1}\frac{\delta F}{\delta y_i}F_i(x,y)$ (여러개 변수에 대한 테일러 급수)를 a에 넣으면
$$ y(x+h)=y(x)+(c_0+c_1)F(x,y)h + c_1\Big[\frac{\delta F}{\delta x}ph+qh\sum_{i=1}^{n-1}\frac{\delta F}{\delta y_i}F_i(x,y) \Big] $$


$ c_0+c_1=1  ~~ ,c_1p=\frac{1}{2} , ~~c_1q=\frac{1}{2} $ 를 구하게 된다.
