$$f'(x) = \frac{f(x+h) - f(x-h)}{2h} + -\frac{h^2}{6}f'''(x) -\cdots$$

$$f'(x) = \frac{f(x+h) - f(x-h)}{2h} + \omicron(h^2)$$

이다.

**From (5):**
$$f''(x) = \frac{f(x+h) - 2f(x)+ f(x-h)}{h^2} + \omicron(h^2)$$

$$f^{(4)}(x) = \frac{f(x+2h) - 4f(x+h)+6f(x)- 4f(x-h) +f(x-2h)}{h^4} + \omicron(h^2)$$

## 4. Richardson Extrapolation

수치 계산의 정확도를 높이는 방법으로 어떤 수량 G를 계산하기 위해 다음과 같이 나타낼 수 있다.

$G = g(h) + E(h)$ 이고 여기서 $g(h)$는 근사식, $E(h)$는 오차 이다.

Richardsom 외삽법은 상수 $c,p$를 정의하고 $E(h) = ch^p$로 둔다.


$h=h_1$인 경우, 
$$G = g(h_1)+ch_1^p ~~~~~~ \rightarrow (1)$$

$h=h_1/2$인 경우,
$$G = g(\frac{h_1}{2})+c\frac{h_1}{2}^p ~~~~~~ \rightarrow (2)$$

오차 항을 제거하기 위해서 $(1)$에서 $2^p \times (2)$를 빼면 $$G=\frac{ g(h_1) - 2^pg(\frac{h_1}{2}) } {1-2^p}$$
 
가 된다.

중앙 차분근사에서 반올림 오차는  $E(h)=O(h^2) = c_1h^2 + c_2h^4 + c_3h^6 + . . . . $ 이 된다.
