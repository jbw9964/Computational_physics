1. How to create a Pandas series from Python lists, ndarrays, and dictionaries.

2. Indexing method for Pandas series.

3. The operations available in a series.


```python
# 파이썬 리스트로부터 판다스 series 만드는 법 
import numpy as np
import pandas as pd

year = [2014,2018,2020,2021] 




```

```python
# numpy array로 series 만들기 
year = np.array([2014,2018,2020,2021])

```

```python
#dictionary로 series 만들기 
graduate = {'철수':2014,'영희':2018,'재석':2020,'명수':2021}

```

```python
# series는 Index와 위치 정보 모두로 Indexing이 가능하다.


# Index 찾기


```

```python
# pandas series는 numpy와 같이 다양한 operation이 가능하다.

ser_1 = pd.Series([1, 2, 3])
ser_2 = pd.Series([4, 5, 6])

# 1. 덧셈/뺄셈

addition = ser_1 + ser_2

# 2.곱셈

multiple = ser_1 * ser_2

# 3. 나눗셈

division = ser_1 / ser_2

print(addition)
print(multiple)
print(division)
```

```python
# Series 끼리의 연산

graduate_1 = {'철수':2014,'영희':2018,'재석':2020,'명수':2021}
graduate_2 = {'철수':1,'영희':2,'미선':3,'명수':4}


graduate_1 = pd.Series(graduate_1)
graduate_2 = pd.Series(graduate_2)

print(graduate_1)
print(graduate_2)
graduate_1.add(graduate_2, fill_value=0)
```
