```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python

x_fit = [1, 2, 3]
y_fit = [3, 2, 0]
print(np.interp(1.35, x_fit, y_fit))    # 보간 시켰을 때, x = 1.35 에 해당하는 y 값
print(np.interp([1, 2, 3, 4], x_fit, y_fit))    # 보간 시켰을 때, x = ... 에 해당하는 y 값들
```

```python
x_plot = np.linspace(0, 10)
y_plot = np.interp(x_plot, x_fit, y_fit)

plt.plot(x_plot, y_plot, label="Interpolation")
plt.plot(x_fit, y_fit, 'o', label="Data point")
plt.legend()
plt.show()
```

```python
x_fit = np.linspace(-np.pi, np.pi)
y_fit = np.sin(x_fit)

x_interp = np.linspace(-np.pi * 4, np.pi * 4, 20)
y_interp = np.sin(x_interp)
y_interp = np.interp(x_interp, x_interp, y_interp)

plt.plot(x_fit, y_fit, label="Function", color='r')
plt.plot(x_interp, y_interp, 'o', label="Interpolation_dot")
plt.plot(x_interp, y_interp, label="Interpolation_line")
plt.legend()
plt.show()

# 함수가 [-pi, pi] 임에도 interp 써서 그 외 범위까지 나타나는 모습
```

```python
# 함수의 모양은 대충 알겠는데, 그 안에 계수를 잘 모르겠는 경우.
# scipy 의 curve_fit 를 쓸 수 있다.
from scipy.optimize import curve_fit

def some_function(x, a, b) : 
    return a * x + b

x_fit = np.array(np.linspace(0, 10, 10))
y_fit = some_function(x_fit, *np.random.random(1), *np.random.random(1))

param, cvar = curve_fit(some_function, x_fit, y_fit)
y_curve_fit = some_function(x_fit, param[0], param[1])
print(param, cvar)
```

```python
plt.plot(x_fit, y_fit, 'o', label="Data_point")
plt.plot(x_fit, y_curve_fit, 'o', label="Curve_fit_dot", alpha=0.4)
plt.plot(x_fit, y_curve_fit, label="Curve_fit_line", alpha=0.5)
plt.legend()
plt.show()
```
